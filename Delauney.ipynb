```python
from scipy.spatial import Delaunay
import numpy as np
import random
N=5
A=10
B=5
points=[]
for i in range(0,N):
    points.append([random.uniform(0,A), random.uniform(0,B)])
a = [[A,0], [A,B], [0,B], [0,0]]
points = np.concatenate((points,a))
tri = Delaunay(points)
import matplotlib.pyplot as plt
plt.triplot(points[:,0], points[:,1], tri.simplices)
plt.plot(points[:,0], points[:,1], 'o')
for j, p in enumerate(points):
     plt.text(p[0]-0.03, p[1]+0.03, j, ha='right')
for j, s in enumerate(tri.simplices):
    p = points[s].mean(axis=0)
    plt.text(p[0], p[1], '*%d' % j, ha='center')
    plt.fill_between(points[s][:,0], points[s][:,1])
plt.show()
P = N + len(a)
T = 2*P - 6
print("Number of points:\n",P)
print("Number of tringles:\n",T)
```
