```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
image = cv2.imread('img.png')
plt.imshow(image)
```

```python
image = cv2.imread('img.png')
image1 = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image1)
```

```python
image2 = cv2.imread('img.png') 
gray_image = cv2.cvtColor(image2,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image) 
```

```python
hsv_image = cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
plt.imshow(hsv_image)
```

```python
image3 = cv2.imread('img.png')
r,c = image3.shape[:2] 
M = np.float32([[1,0,-150],[0,1,-150]]) 
dst = cv2.warpAffine(image3,M,(c,r)) 
plt.imshow(dst)
```

```python
def px(size, proportion):
    return int(size * proportion)
image4 = image[px(r, .25):px(r, .50), 
          px(c, .30):px(c, .40)]
plt.imshow(image4)
```

```python
image = cv2.imread('img.png',0)
ret,thresh_binary = cv2.threshold(image,127,255,cv2.THRESH_BINARY)
image5 = thresh_binary
plt.imshow(image5,'gray')
```
