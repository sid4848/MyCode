```python
from pylab import plot
import numpy as np
import matplotlib.pyplot as plt

def f(x): # sample function
    return np.sin(np.pi*x)

# evaluation of the function
x = np.linspace(0,1,150)
y = f(x)

b = np.linspace(0,1,30)
for a in b:
    h = 0.1
    fprime = (f(a+h)-f(a))/h 
    tan = f(a)+fprime*(x-a)  
    plot(x,y,'w',a,f(a),x,tan,'r')
plt.show()
```
