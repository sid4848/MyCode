```python
import numpy as np
x = np.linspace(0, 5, num=50)
y = 1 /(1 + 9*np.exp(-2*x))
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(x,y,'r')
plt.xlabel('time')
plt.ylabel('fraction growth')
plt.title('Evalution of natural phenomenon')
plt.savefig('EONP.png', dpi=300)
```
